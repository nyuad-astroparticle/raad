```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
# import the data
raw = pd.read_csv('./storage/pc_buff4/21.csv')
```

```python
ent = raw['entry_nr'].to_numpy()        # Get the entry numbers
idx = np.argsort(ent)                   # Returns the indices of the sorted array

# Plot the stuffs
fig = plt.figure()
plt.scatter(range(len(ent)),ent,s=0.1)
plt.scatter(range(len(ent)),ent[idx],s=0.1)
```
